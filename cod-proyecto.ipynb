```python
r = open("./columnas_inca.txt", "r")

columnas_inca = []
columnas_adni = []

for line in r:
    new_line_list = line.split("=")
    x = []
    for e in range(len(new_line_list)):
        x.append(new_line_list[e].strip())
    columnas_inca.append(x[0])
    columnas_adni.append(x[1])

print(columnas_adni)
print(" ")
print(len(columnas_inca))   
```
